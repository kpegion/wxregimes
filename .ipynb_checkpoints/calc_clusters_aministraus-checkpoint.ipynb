```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
from array import array

import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker


from sklearn.cluster import KMeans
```

```python
path='/project/mjo/straus/ERAI_T42/eofs/DJF/5day_means/'
fname_pcs='pcs_Z500U250_PNA_DJF_T42.1980-2014.dat'
fname_eofs='eofs_Z500U250_PNA_DJF_T42.1980-2014.dat'
```

```python
# From .ctl file provided by David
x_nx=1
y_nyrs=35
z_npcs=50
t_months=88
```

```python
# fortran record length for sequential access
reclfseq=(x_nx*y_nyrs+2)*4
```

```python
# Open file
luin = open(path+fname_pcs,'rb')

# Create empty numpy array
data=np.zeros((y_nyrs,x_nx,t_months,z_npcs))

# Read each record and store in data
for t in range(t_months):
    for z in range(z_npcs):
        tmp=luin.read(reclfseq) # read in fortran record in bytes, you will have to do consectivte reads to get other variables
        tmp1=array('f',tmp)  # convert to single precision (real 32bit)
        tmp2=tmp1[1:-1] # pull out data array (leaving behind fortran control records)  for fortran sequential
        data[:,:,t,z]=np.reshape(tmp2,(y_nyrs,x_nx))  # create a 2d array
        
# Get rid of x=1 dimension and reshape to time x pcs
data=np.squeeze(data).reshape(y_nyrs*t_months,z_npcs)

# Check shape
data.shape
```

### Read PCs

```python
nt=y_nyrs*t_months
npcs=z_npcs

ds=xr.DataArray(data,
                coords={'time':np.arange(nt),
                        'pcnum':np.arange(npcs)},
                        dims=['time','pcnum'])        
ds=ds.to_dataset(name='pcs')
#ds=ds.transpose()
ds=ds.sel(pcnum=slice(0,11))

ds
```

```python
plt.plot(ds['pcs'][0:89,0])
```

### Apply K-Means Clustering

```python
#kmeans = KMeans(n_clusters=4,init='random',random_state=1,n_init=100)
#kmeans = KMeans(n_clusters=4,init='random',random_state=1)
kmeans = KMeans(n_clusters=4,init='random',random_state=1,n_init=100)
kmeans.fit(ds['pcs'])
y_kmeans=kmeans.predict(ds['pcs'])
```

```python
plt.scatter(ds['pcs'][:, 0], ds['pcs'][:, 1], 
            c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], 
            c='black', s=200, alpha=0.5);
```
