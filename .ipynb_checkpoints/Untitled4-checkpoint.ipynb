```python
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fname='/shared/working/rean/era-interim/daily/data/1980/ei.oper.an.pl.regn128cm.1980120300'
```

```python
# Read Geopotential Height (Z) data           
ds=xr.open_dataset(fname,engine='cfgrib',
                        backend_kwargs={'indexpath':'',
                                        'filter_by_keys':{'name': 'Geopotential'}})
```

```python
ds
```

```python
ds_pna=ds.sel(isobaricInhPa=500,latitude=slice(80,20),longitude=slice(150,300))
```

```python
ds_pna
```

```python
plt.contourf((ds_pna['z'][::-1,:])/9.81)
plt.colorbar()
```

```python
# Interpolate data to coarser grid
ds_new_grid=xr.open_dataset('tempgaussian.128x64.nc')
ds_new_grid
```

```python
ds_new_grid_pna=ds_new_grid.rename({'temp':'z'}).sel(lat=slice(80,20),lon=slice(150,300))
ds_new_grid_pna
```

```python
ds_regrid=ds_pna.interp_like(ds_new_grid_pna)
```

```python
ds_regrid
```

```python
plt.contourf((ds_regrid['z'][::-1,:])/9.81)
```
