```python
yrs_list=np.arange(1981,2016)
varname='z500'
da=ds_reind[varname]
window=5
nt,ny,nx=da.shape
ndays=90
nyr=int(nt/ndays)

ds_rm_tmp=[]
    
for yr in yrs_list:
        
    start_date=str(yr-1)+'1201'
    end_date=str(yr)+'0228'
        
    ds_this_year=da.sel(time=slice(start_date,end_date))
    
    tmp=ds_this_year.rolling(time=window,center=True).mean().dropna(dim='time')
    ds_rm_tmp.append(tmp)

#    plt.plot(ds_this_year['time'],ds_this_year[:,12,25])
    plt.plot(tmp['time'],tmp[:,12,25])
    
ds_rm = xr.combine_nested(ds_rm_tmp, concat_dim='time').to_dataset(name=varname)
ds_rm
```

```python
ds_rm
```

```python
plt.plot(ds_rm['time'],ds_rm['z500'][:,12,25])
```

```python
da_new=ds_rm['z500'].values

nt,ny,nx=da_new.shape
print(da_new.shape)
ndays=86
nyr=int(nt/ndays)
print(nyr, ndays, ny, nx)    
    
poly=np.zeros((nyr,ndays,ny,nx))
x=np.arange(ndays)
y=da_new.reshape(nyr,ndays,nx*ny)
ytmp=y.reshape(nyr,ndays,ny,nx)

for iyr in range(nyr):
    
    pfitcoeffs=np.polyfit(x,y[iyr,:,:],2)
    
    k=0
    for j in range(ny):
        for i in range(nx):
            poly[iyr,:,j,i]=np.polyval(pfitcoeffs[:,k],x)
            k+=1
#    plt.plot(poly[iyr,:,12,25])
climo=poly.mean(axis=0)

plt.plot(ytmp[30,:,12,25])
plt.plot(poly[30,:,12,25])
plt.plot(climo[:,12,25],color='k')
```

```python
da_new=ds_rm['z500'].values

nt,ny,nx=da_new.shape
ndays=86
nyr=int(nt/ndays)

print(nyr,ndays,ny,nx)

y=da_new.reshape(nyr,ndays,ny,nx)
anoms=np.zeros((nyr,ndays,ny,nx))

for iyr in range(nyr):
    anoms[iyr,:,:,:]=y[iyr,:,:,:]-climo 
#    plt.plot(y[iyr,:,12,25],color='b') 
#    plt.plot(climo[:,12,25],color='k')
    plt.plot(anoms[iyr,:,12,25])
anoms=anoms.reshape(nt,ny,nx)
```

```python
plt.plot(anoms[:,12,25])
```

```python
ds_z500_anoms=xr.DataArray(anoms,
                coords={'time':ds_rm['time'],
                        'lat':ds_rm['lat'],
                        'lon': ds_rm['lon']},
                        dims=['time','lat','lon'])        
ds_z500_anoms=ds_z500_anoms.to_dataset(name='z500')
```

```python
coslat = np.cos(np.deg2rad(ds['lat'].values))
weights_array = np.sqrt(coslat)[:, np.newaxis]
```

```python
ds_z500_norm=ds_z500_anoms['z500']*weights_array
norm_z500=np.sqrt((ds_z500_norm.var(dim='time')).mean(dim=['lat','lon']))
ds_z500_norm=ds_z500_norm/norm_z500
```

```python
norm_z500*norm_z500
```

```python
from eofs.standard import Eof
```

```python
solver = Eof(ds_z500_norm.values)
```

```python
eof1 = solver.eofsAsCovariance(neofs=1)
```

```python
pcs = solver.pcs(npcs=1)
```
