````python
import numpy as np
from array import array
import xarray as xr
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
````

````python
path='/project/mjo/straus/ERAI_T42/eofs/DJF/5day_means/'
fname_pcs='pcs_Z500U250_PNA_DJF_T42.1980-2014.dat'
fname_eofs='eofs_Z500U250_PNA_DJF_T42.1980-2014.dat'
````

### Read the PCs file

````python
# From .ctl file provided by David
x_nx=1
y_nyrs=35
z_npcs=50
t_months=88
````

````python
# fortran record length for sequential access
reclfseq=(x_nx*y_nyrs+2)*4
````

````python
# Open file
luin = open(path+fname_pcs,'rb')

# Create empty numpy array
data=np.zeros((y_nyrs,x_nx,t_months,z_npcs))

# Read each record and store in data
for t in range(t_months):
    for z in range(z_npcs):
        tmp=luin.read(reclfseq) # read in fortran record in bytes, you will have to do consectivte reads to get other variables
        tmp1=array('f',tmp)  # convert to single precision (real 32bit)
        tmp2=tmp1[1:-1] # pull out data array (leaving behind fortran control records)  for fortran sequential
        data[:,:,t,z]=np.reshape(tmp2,(y_nyrs,x_nx))  # create a 2d array
        
# Get rid of x=1 dimension and reshape to time x pcs
data=np.squeeze(data).reshape(y_nyrs*t_months,z_npcs)

# Check shape
data.shape
````

````python
data=np.flip(data,axis=1) # is this right?
````

### Read the EOFs File

````python
# From .ctl file provided by David
x_nlons=128
y_nlats=64
t_neofs=50
n_vars=2
````

````python
# fortran record length for sequential access
reclfseq=(x_nlons*y_nlats+2)*4
````

````python
# Open file
luin = open(path+fname_eofs,'rb')


# Create empty numpy array
eofs=np.zeros((t_neofs,y_nlats,x_nlons,n_vars))

# Read each record and store in data
for t in range(t_neofs):
    for v in range(n_vars):
        tmp=luin.read(reclfseq) # read in fortran record in bytes, you will have to do consectivte reads to get other variables
        tmp1=array('f',tmp)  # convert to single precision (real 32bit)
        tmp2=tmp1[1:-1] # pull out data array (leaving behind fortran control records)  for fortran sequential
        eofs[t,:,:,v]=np.reshape(tmp2,(y_nlats,x_nlons))  # create a 2d array
        
# Extract z500 and u250
eof_z500=eofs[:,:,:,0]
eof_u250=eofs[:,:,:,1]
````

````python
lats=[-87.86, -85.10, -82.31, -79.53, -76.74, -73.95, -71.16, -68.37, \
      -65.58, -62.79, -60.00, -57.21, -54.42, -51.63, -48.84, -46.04, \
      -43.25, -40.46, -37.67, -34.88, -32.09, -29.30, -26.51, -23.72, \
      -20.93, -18.14, -15.35, -12.56, -9.77,  -6.98,  -4.19,  -1.40, \
      1.40,   4.19,   6.98,   9.77,  12.56,  15.35,  18.14,  20.93, \
      23.72,  26.51,  29.30,  32.09,  34.88,  37.67,  40.46,  43.25, \
      46.04,  48.84,  51.63,  54.42,  57.21,  60.00,  62.79,  65.58, \
      68.37,  71.16,  73.95,  76.74,  79.53,  82.31,  85.10,  87.86]
lats=lats[::-1]
print(lats)
lons=np.arange(128)*2.8125 + 0.0
eof_mode=np.arange(t_neofs)
````

### Make DataArray for z500

````python
eof_z500=xr.DataArray(eof_z500,
                coords={'eofs':eof_mode,
                        'lat':lats,
                        'lon': lons},
                        dims=['eofs','lat','lon'])        
eof_z500=eof_z500.to_dataset(name='z500')
eof_z500
````

### Make DataArray for u250

````python
eof_u250=xr.DataArray(eof_u250,
                coords={'eofs':eof_mode,
                        'lat':lats,
                        'lon': lons},
                        dims=['eofs','lat','lon'])        
eof_u250=eof_u250.to_dataset(name='u250')
eof_u250
````

### Select data only for PNA Region [20-80; 150-300]

````python
eof_z500_pna=eof_z500.sel(lat=slice(80,20),lon=slice(150,300))
eof_u250_pna=eof_u250.sel(lat=slice(80,20),lon=slice(150,300))
eof_z500_pna
````

### Make Various test plots

````python
cm = 180
proj = ccrs.PlateCarree(central_longitude=cm)
minlon = -30 + cm
maxlon = +120 + cm
lat_labels=np.arange(20,90,20)
lon_labels=np.arange(150,330,30)
extent=[minlon,maxlon,20,80]
````

````python
ax=plt.axes(projection=proj)
ax.set_extent(extent,crs=ccrs.PlateCarree())
cs=ax.contourf(eof_z500_pna['lon'],eof_z500_pna['lat'],
               eof_z500_pna['z500'][0,:,:],transform=ccrs.PlateCarree())
ax.set_title('EOF1 Z500')
ax.coastlines()
plt.colorbar(cs)
````

````python
ax=plt.axes(projection=proj)
ax.set_extent(extent,crs=ccrs.PlateCarree())
cs=ax.contourf(eof_u250_pna['lon'],eof_u250_pna['lat'],
               eof_u250_pna['u250'][0,:,:],transform=ccrs.PlateCarree())
ax.set_title('EOF1 u250')
ax.coastlines()
plt.colorbar(cs)
````

````python
ax=plt.axes(projection=proj)
ax.set_extent(extent,crs=ccrs.PlateCarree())
cs=ax.contourf(eof_z500_pna['lon'],eof_z500_pna['lat'],
               eof_z500_pna['z500'][1,:,:],transform=ccrs.PlateCarree())
ax.set_title('EOF2 Z500')
ax.coastlines()
plt.colorbar(cs)
````
